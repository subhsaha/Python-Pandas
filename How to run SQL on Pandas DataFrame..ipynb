```python
import pandas as pd
import pandasql as ps
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
q1 = """SELECT * FROM ufo a join ufo b on a.city = b.city limit 10"""
```

```python
ps.sqldf(q1, locals())
```

#### Steps detailing ?    

* import options in python is used to import the packages.On the above statement two packages are imported and is renamed as pd and ps for easy use.
